```python
#data cleaning methods
data = pd.read_csv("titanic.csv")
```

```python
data.isnull().sum()
```

```python
null_pct = data.apply(pd.isnull).sum()/data.shape[0]
```

```python
null_pct
```

```python
valid_columns = data.columns[null_pct < .30]
```

```python
valid_columns
```

```python
for i in data.columns:
    if(data[i].isnull().sum()>int((418/100)*30)):
       data.drop(i,axis=1,inplace=True)
```

```python
data
```

```python
#imputing Method
data["Age"].median(), data["Age"].mean()
```

```python
data["Age"].fillna(data["Age"].mean(), inplace=True)
```

```python
data.isna().sum()
```

```python
data["Fare"].median(), data["Fare"].mean()
```

```python
data["Fare"].fillna(data["Fare"].mean(), inplace=True)
```
